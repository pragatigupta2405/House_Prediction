```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_excel('/content/housingfile.xlsx')
data.head()
```

```python
data.isnull().sum()
```

```python
data['furnishingstatus'].value_counts()
```

```python
data1=data.copy()
```

```python
data1.shape
```

```python
data.shape
```

```python
data1['furnishingstatus'].replace({'furnished':2,'semi-furnished':1,'unfurnished':0},inplace=True)
data1['airconditioning'].replace({'yes':1,'no':0},inplace=True)
data1['mainroad'].replace({'yes':1,'no':0},inplace=True)
data1['guestroom'].replace({'yes':1,'no':0},inplace=True)
data1['basement'].replace({'yes':1,'no':0},inplace=True)
data1['parking'].replace({'yes':1,'no':0},inplace=True)
data1['prefarea'].replace({'yes':1,'no':0},inplace=True)

```

```python
data1.head()
```

```python
data1['hotwaterheating'].replace({'yes':1,'no':0},inplace=True)
```

```python
data1.head()
```

```python
X=data1.drop('price',axis=1)
y=data1['price']
```

```python
print(X)
```

```python
data1.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
y_pred_score=model.score(X_test,y_test)
print(y_pred_score)
#
```

```python
x_train_pred=model.predict(X_train)
train_data_score=model.score(X_train,y_train)
print(train_data_score)
```

```python
r2_score(y_test,y_pred)
```
